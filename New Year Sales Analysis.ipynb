Exploratory Data Analysis

```python
# Gender Analysis
ax=sns.countplot(x = 'Gender',data = nw)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = nw.groupby(['Gender'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)
sns.barplot(x = 'Gender',y = 'Amount',data = sales_gen)          
```

From the above chart we can see that most of the buyers are Female and even the purchasing power of females are greater than men

```python
# Age group analysis
ax=sns.countplot(x = 'Age Group',data = nw,hue ='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Total sales vs age group
sales_age = nw.groupby(['Age Group'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)
sns.barplot(x = 'Age Group',y = 'Amount',data = sales_age)  
```

According to age group from the chart buyers between the age groups are 26 - 35 has highest purchasing power 

```python
#State wise analysis
# Total number of orders from the top 10 states 
sales_state = nw.groupby(['State'],as_index = False)['Orders'].sum().sort_values(by = 'Orders',ascending = False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x = 'State',y = 'Orders',data = sales_state)  
```

```python
#Total amount of sales from top 10 states
sales_state = nw.groupby(['State'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x = 'State',y = 'Amount',data = sales_state)  
```

From above graphs we can see that unexpectedly most of the orders are from Uttar Pradesh,Maharastra,Karnataka and it has the purchasing orders

```python
#Marital Status Analysis
ax=sns.countplot(x = 'Marital_Status',data = nw)
sns.set(rc={'figure.figsize':(3,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_marital = nw.groupby(['Marital_Status','Gender'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(x = 'Marital_Status',y = 'Amount',data = sales_marital,hue = 'Gender') 
```

From above graphs we can see that most of the buyers are married(women) and they have high purchasing power

```python
# Occupation Analysis
ax=sns.countplot(x = 'Occupation',data = nw)
sns.set(rc={'figure.figsize':(20,10)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_occ = nw.groupby(['Occupation'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)
sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(x = 'Occupation',y = 'Amount',data = sales_occ) 
```

IT,Aviation and Health care sector have the most buyers

```python
#Product Category Analysis
ax=sns.countplot(x = 'Product_Category',data = nw)
sns.set(rc={'figure.figsize':(20,10)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_pdt = nw.groupby(['Product_Category'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False).head(10)
sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(x = 'Product_Category',y = 'Amount',data = sales_pdt) 
```

From the above data frame observation the top contributors we can 
see that Food, Clothing, and Electronics appears prominently, 
that likely means these are the most frequently purchased or have highest total spending across age groups 26 - 35 and genders (Female)
